```python
# 找到某個詞出現的頻率（百分比）
100*text4.count('a')/len(text4)
```

```python
# 平均詞密度
def lexital_diversity(text):
    return len(text)/len(set(text))
```

```python
def percentage(count,total):
    return 100 * count / total
```

```python
lexital_diversity(text3)
```

```python
lexital_diversity(text5) #詞的重複率沒有text3高
```

```python
percentage(4,5)
```
