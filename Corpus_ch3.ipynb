```python
#計算統計文本
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))#包含空白字符(raw函數回傳沒有進行任何語言處理的內容)
    num_words = len(gutenberg.words(fileid))#平均詞長
    num_sents = len(gutenberg.sents(fileid))#平均句長
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)]))#每個詞出現的平均次數（多樣性）
    print (int(num_chars/num_words), int(num_words/num_sents), int(num_words/num_vocab), fileid)
```

```python
def generate_model(cfdist, word, num=15):
    for i in range(num):
        print(word)
        word = cfdist[word].max()
```

```python
text = nltk.corpus.genesis.words('english-kjv.txt')
```

```python
bigrams = nltk.bigrams(text)
```

```python
cfd = nltk.ConditionalFreqDist(bigrams)
```

```python
print (cfd['living'])
```

```python
generate_model(cfd, 'living')
```

```python
'''cfdist= ConditionalFreqDist(pairs) 从配对链表中创建条件频率分布
cfdist.conditions() 将条件按字母排序
cfdist[condition] 此条件下的频率分布
cfdist[condition][sample] 此条件下给定样本的频率
cfdist.tabulate() 为条件频率分布制表
cfdist.tabulate(samples, conditions) 指定样本和条件限制下制表
cfdist.plot() 为条件频率分布绘图
cfdist.plot(samples, conditions) 指定样本和条件限制下绘图
cfdist1 < cfdist2 测试样本在cfdist1 中出现次数是否小于在cfdist2 中出现次數'''
```

```python
#複數形式的轉換
def plural(word):
    if word.endswith('y'):
        return word[:-1] + 'ies'
    elif word[-1] in 'sx' or word[-2:] in ['sh', 'ch']:
        return word + 'es'
    elif word.endswith('an'):
        return word[:-2] + 'en'
    else:
        return word + 's'
```

```python
plural('fairy')
```

```python
plural('woman')
```

```python
# 過濾文本：計算文本的詞彙表，然後只留下罕見或拼寫錯誤的詞
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)
```

```python
unusual_words(nltk.corpus.nps_chat.words())
```

```python
#停用詞語料庫
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
#計算文本中沒有在停用詞語料庫的比例
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)
```

```python
content_fraction(nltk.corpus.reuters.words()) #篩選掉文本中三分之一的詞
```
