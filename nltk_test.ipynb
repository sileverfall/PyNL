```python
# 重複詞的密度
from __future__ import division
len(text3)/len(set(text3)) #平均一個詞出現幾次
```

```python
# 平均詞密度
def lexital_diversity(text):
    return len(text)/len(set(text))
```

```python
def percentage(count,total):
    return 100 * count / total
```
